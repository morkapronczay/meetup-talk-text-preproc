```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
google_trends = pd.read_csv('input/GoogleTrends.csv', skiprows=2)
google_trends.columns = ['Date', 'NLP', 'Text-mining', 'Text-analysis']
google_trends['Date'] = [f.date() for f in pd.to_datetime(google_trends['Date']).tolist()]
```

```python
fig = google_trends.set_index('Date').plot(figsize=(20,10), linewidth=3).get_figure()
```

```python
fig.savefig('outputs/google-trends.png')
```
